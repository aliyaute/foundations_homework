```python
cucumber = []

for page_num in range(1, 11):
    url = (f"https://www.epicurious.com/search/cucumbers?page={page_num}")
    #print("Now scraping", url)

    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    
    items = doc.find_all("article", class_="recipe-content-card")
    for item in items: 
        row = {}
        try:
            row['Tag'] = item.find(class_='tag').text.strip()
            row['Title'] = item.find('a').get_text()
            row['Would Make Again'] = item.find(class_="make-again-percentage").text
            row['Summary'] = item.find(class_="summary").text.strip()
            row['Link'] = item.find('a')['href']
            row['Rating'] = item.find(class_="rating").text.strip()
        
        except:
            pass

        cucumber.append(row)
            
        #print(cucumber)
    
df = pd.DataFrame(cucumber)
df.head()        
```

```python
df = pd.DataFrame(cucumber)
df.to_csv("cucumber.csv", index=False)
df.head(10)

 
    
```

```python
def scrape_page(row):
    url = f"https://www.epicurious.com{row['Link']}"
    print = ("Scraping", url)
    
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    
    try:
        food = {}
        food['Ingredients'] = doc.find(class_="ingredient-groups").text.strip()
        food['Directions'] = doc.find(class_='preparation-groups').text.strip()
        food['Tags'] = doc.find(class_="tags").text
        
        return pd.Series(food)

    except: 
        pass
```

```python
scrape_page(df.loc[2])
```

```python
scraped_df = df.apply(scrape_page, axis=1)
scraped_df
```
