```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://www.goodreads.com/list/show/6934.Science_Fiction_Books_by_Female_Authors")
doc = BeautifulSoup(response.text)
```

```python
items = doc.find_all("tr")
```

```python
books = []

for item in items:
    row = {}
    
    row['Rank'] = item.find('td', class_='number').text.strip()
    row['Title'] = item.find(class_='bookTitle').text.strip()
    row['Author'] = item.find(class_='authorName').text.strip()
    row['Score'] = item.find(class_='smallText uitext').find('a').text.strip()
    row['Votes'] = item.find_all('a')[4].text.strip()  
    row['Ratings'] = item.find(class_="minirating").text.strip()

    books.append(row)
    


```

```python
import re
```

```python
books_list = books
books_list
```

```python
df=pd.DataFrame(books)
df.head()
```

```python
df.to_csv("books.csv", index=False)
```

```python
df['Series'] = df.Title.str.extract("\((.*),")
df['Number in Series'] = df.Title.str.extract("\(.*,\s[#](.)\)")
df.Title = df.Title.str.extract("([^\(]+)")
df['Number of Ratings'] = df.Ratings.str.extract("—\s(.*)\s")
df.Ratings = df.Ratings.str.extract("(\d.\d+)")
```
